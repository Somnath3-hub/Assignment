Q 1 .  What is the difference between AI, ML, DL, and Data Science? Provide a 
brief explanation of each. 
(Hint: Compare their scope, techniques, and applications for each.)

->    Artificial Intelligence (AI) is the broadest concept in the field of computer science. It aims to create machines that can perform tasks that normally require human intelligence — like thinking, reasoning, learning, and decision-making. AI systems are designed to mimic human behavior and logic.
For example, chatbots, voice assistants (like Alexa), and self-driving cars are applications of AI.

Machine Learning (ML) is a subset of AI. It focuses on developing algorithms that can learn automatically from data. Instead of being explicitly programmed, ML models learn patterns and make predictions or classifications based on experience.
For example, spam email detection, movie recommendations, or price prediction are examples of ML applications.

Deep Learning (DL) is a specialized branch of ML that uses artificial neural networks with multiple layers (hence “deep”). DL algorithms automatically extract features from raw data such as images, audio, or text.
Examples include image recognition, speech-to-text systems, and face detection. DL requires large datasets and high computational power (GPUs).

Data Science is a broader discipline that combines statistics, programming, and domain knowledge to extract useful insights from data. Data Science often uses AI and ML techniques to analyze, visualize, and interpret large amounts of data for decision-making.
Examples: Business analytics, fraud detection, and customer segmentation.

Q 2. Explain overfitting and underfitting in ML. How can you detect and prevent 
them? 
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

-> In Machine Learning, creating a model that performs well on both training and testing data is important. Two major problems that affect model performance are Overfitting and Underfitting.

    
Overfitting:
Overfitting occurs when a model learns the training data too well — it captures not only the actual patterns but also random noise or fluctuations. As a result, the model performs excellently on the training dataset but fails on new unseen data.
Detection: Training accuracy is very high, but test accuracy is much lower.

Prevention:

1.Use cross-validation to check generalization.

2.Apply regularization (L1 or L2) to reduce model complexity.

3.Use dropout in neural networks.

4.Simplify the model by reducing layers or parameters.

5.Gather more diverse data.


Underfitting:
Underfitting happens when a model is too simple to learn the underlying structure of the data. It performs poorly on both training and test data.
Detection: Both training and test accuracy are low.

Prevention:

1.Increase model complexity (more features or layers).

2.Train longer with proper hyperparameters.

3.Reduce regularization if it’s too strong.

    
Summary:
Overfitting → High variance, Low bias (model memorizes data).
Underfitting → High bias, Low variance (model fails to learn).
A well-balanced model generalizes well to unseen data.

Q 3. How would you handle missing values in a dataset? Explain at least three 
methods with examples. 
Hint: Consider deletion, mean/median imputation, and predictive modeling.

->  Missing values are common in real-world datasets. They can appear due to data entry errors, incomplete surveys, or technical issues during data collection. If not handled properly, they can reduce accuracy and lead to biased analysis.

There are several methods to handle missing values:


1.Deletion Methods:

Row deletion: If the number of missing rows is very small (less than 5%), we can remove those records.

Column deletion: If a column has too many missing values, we can drop that column completely.

Example: If only 2% of the rows have missing “Age”, remove those rows.

2.Imputation using Mean, Median, or Mode:

For numeric features: use mean or median to replace missing values.

For categorical features: use mode (most frequent category).

Example: If Age is missing, replace it with the average age.

3.Predictive Imputation:

Use another model (like linear regression or KNN) to predict missing values using other variables.

Example: Predict missing salary based on education and experience.

4.KNN Imputer or Interpolation:

Fills missing data using nearest neighbors or previous time-based values.


Handling missing data properly ensures cleaner datasets and better-performing models.

Q 4. What is an imbalanced dataset? Describe two techniques to handle it 
(theoretical + practical). 
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.

-> An imbalanced dataset occurs when one class has a much larger number of records compared to another class. For example, in fraud detection datasets, 98% of transactions are non-fraud, while only 2% are fraud.

    
Problem:
The model becomes biased towards the majority class and fails to learn the characteristics of the minority class. This can result in high accuracy but poor real-world performance.

Techniques to handle imbalance:

    1.Oversampling:

Increases the number of samples in the minority class.

SMOTE (Synthetic Minority Over-sampling Technique) generates synthetic examples using interpolation between existing minority samples.

Helps balance the dataset without simple duplication.

    2.Undersampling:

Reduces the number of samples in the majority class.

Helps when dataset size is large but may lose useful information.

    3.Class Weight Adjustment:

Assigns higher weight to minority class samples during training so that errors on them are penalized more.

    
Evaluation Metrics:

In imbalanced datasets, accuracy is not a reliable metric. Instead, use Precision, Recall, F1-Score, and ROC-AUC to measure true model performance.

Q 5. Why is feature scaling important in ML? Compare Min-Max scaling and 
Standardization. 
Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient 
descent.

->  Feature scaling is the process of adjusting the values of numerical features so they are on a similar scale. It ensures that one feature does not dominate others due to its range or units.

Why it is important:
Many ML algorithms (like KNN, SVM, Logistic Regression, and Gradient Descent) rely on distance calculations or gradients. If features have very different ranges, the algorithm may behave incorrectly. Scaling improves both accuracy and training speed.

    
Types of Scaling:

    1.Min-Max Scaling (Normalization):

Formula: (x - min) / (max - min)

Scales all values to a fixed range between 0 and 1.

Preserves data shape but changes magnitude.

Useful for neural networks and algorithms sensitive to range.

    2.Standardization (Z-score Normalization):

Formula: (x - mean) / standard deviation

Transforms data to have mean = 0 and standard deviation = 1.

Works well with algorithms like SVM, Logistic Regression, and PCA.

    
Example:
If “Age” = [20, 40, 60] and “Salary” = [20,000, 200,000, 500,000],
the model might overvalue salary. Scaling balances them, giving fair importance to both features.

Q 6. Compare Label Encoding and One-Hot Encoding. When would you prefer 
one over the other? 
Hint: Consider categorical variables with ordinal vs. nominal relationships.

->  Machine Learning algorithms require numerical input, but many datasets contain categorical data (like gender, color, or city). Encoding techniques convert these categorical features into numeric form.

1. Label Encoding:

Assigns a unique number to each category.

Used for ordinal data (when order matters).

Example: Education → School = 0, College = 1, Graduate = 2.

Advantages: Simple, efficient, and compact.

Disadvantages: Model may assume that higher values mean higher importance, which may not be true.

2. One-Hot Encoding:

Creates binary columns for each category.

Used for nominal data (no order).

Example: Color → Red = [1,0,0], Blue = [0,1,0], Green = [0,0,1].

Advantages: Prevents false ordering among categories.

Disadvantages: Increases the number of columns, which can make models slower for large datasets.

When to use:

Label Encoding → Ordered categories like size (Small, Medium, Large).

One-Hot Encoding → Unordered categories like gender or city.


Conclusion:
Both encoding methods are essential preprocessing steps in machine learning, helping models understand categorical data effectively.

Q 7.  Google Play Store Dataset 
a). Analyze the relationship between app categories and ratings. Which categories have the 
highest/lowest average ratings, and what could be the possible reasons? 
Dataset: https://github.com/MasteriNeuron/datasets.git

->

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("googleplaystore.csv", encoding='latin-1')

# Step 2: Clean the data (remove missing or invalid ratings)
df = df.dropna(subset=['Rating'])
df = df[df['Rating'] <= 5]  # keep only valid ratings

# Step 3: Group by category and calculate average rating
category_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Step 4: Display top and bottom categories
print("Top 5 categories with highest average ratings:\n", category_rating.head(5))
print("\nBottom 5 categories with lowest average ratings:\n", category_rating.tail(5))

# Step 5: Visualization
plt.figure(figsize=(12,6))
category_rating.plot(kind='bar', color='skyblue')
plt.title("Average Rating by App Category", fontsize=14)
plt.xlabel("App Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.show()
```

Explanation :
After analyzing the Google Play Store dataset, I found that the average rating of apps differs across various categories.

Highest Rated Categories: 
 Education, Books & Reference, Events, and Art & Design have the highest average ratings.
These categories generally contain useful, learning-based, or creative apps, which attract more positive user feedback because users find them educational and less buggy.

Lowest Rated Categories:
 Action Games, Entertainment, Finance, and Social apps have lower ratings.
This may be because these apps often face performance issues, frequent ads, or bugs. Also, in gaming and social apps, user expectations are very high, which leads to more critical ratings.

Conclusion:
 App categories focusing on learning and creativity usually receive higher ratings, while entertainment and gaming apps face lower ratings due to competition and higher user expectations.

Q 8. Titanic Dataset 
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest 
survival rate, and why do you think that happened? 
b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and 
adults (Age ≥ 18). Did children have a better chance of survival? 
Dataset: https://github.com/MasteriNeuron/datasets.git 

-> 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset (file should be in same folder as notebook)
df = pd.read_csv("titanic.csv")

# --- Survival by Pclass ---
df_class = df.dropna(subset=['Survived','Pclass']).copy()
df_class['Pclass'] = df_class['Pclass'].astype(int)

survival_by_class = df_class.groupby('Pclass')['Survived'].mean().sort_index()
count_by_class = df_class.groupby('Pclass')['Survived'].size()
print("Survival rate by Pclass:")
print(pd.concat([survival_by_class.rename('survival_rate'), count_by_class.rename('count')], axis=1))

plt.figure(figsize=(6,4))
plt.bar(survival_by_class.index.astype(str), survival_by_class.values)
plt.title("Survival Rate by Passenger Class (Pclass)")
plt.xlabel("Pclass")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
for i, v in enumerate(survival_by_class.values):
    plt.text(i, v+0.03, f"{v:.2f}", ha='center')
plt.show()

# --- Survival by Age group (Child < 18, Adult >= 18) ---
df_age = df.dropna(subset=['Survived','Age']).copy()
df_age['Age'] = pd.to_numeric(df_age['Age'], errors='coerce')
df_age = df_age.dropna(subset=['Age'])
df_age['AgeGroup'] = np.where(df_age['Age'] < 18, 'Child', 'Adult')

survival_by_agegroup = df_age.groupby('AgeGroup')['Survived'].mean().reindex(['Child','Adult'])
count_by_agegroup = df_age.groupby('AgeGroup')['Survived'].size()
print("\nSurvival rate by Age Group (Child <18 vs Adult >=18):")
print(pd.concat([survival_by_agegroup.rename('survival_rate'), count_by_agegroup.rename('count')], axis=1))

plt.figure(figsize=(5,4))
plt.bar(survival_by_agegroup.index, survival_by_agegroup.values)
plt.title("Survival Rate: Children vs Adults")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.ylim(0,1)
for i, v in enumerate(survival_by_agegroup.values):
    plt.text(i, v+0.03, f"{v:.2f}", ha='center')
plt.show()
```

Explanation :
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?

I calculated the survival rate for each passenger class in the Titanic dataset. The results show that 1st class passengers had the highest survival rate (≈ 62.96%), 2nd class passengers had a moderate survival rate (≈ 47.28%), and 3rd class passengers had the lowest survival rate (≈ 24.24%).

Possible reasons:

Access and location: First-class cabins were closer to decks and lifeboats, so those passengers could reach safety faster.

Priority and assistance: Crew and rescue operations likely assisted wealthier or first-class passengers more quickly.

Crowding and barriers: Third-class passengers were often below decks and faced crowding or locked/blocked access that delayed evacuation.
    

b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival?

I grouped passengers into Child (Age < 18) and Adult (Age ≥ 18) and computed survival rates. The analysis shows children had a higher survival rate (≈ 53.98%) compared to adults (≈ 38.10%). This supports the historical understanding that during evacuation, priority was given to children (and women), often summarized as “women and children first.” Therefore, children indeed had a better chance of survival.


Conclusion (short):
First-class passengers and children had notably higher survival rates. The main reasons are likely a combination of location on the ship, evacuation priorities, and logistical barriers that affected lower-class passengers.

Q 9. Flight Price Prediction Dataset 
a) How do flight prices vary with the days left until departure? Identify any exponential price 
surges and recommend the best booking window. 
b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are 
consistently cheaper/premium, and why? 
Dataset: https://github.com/MasteriNeuron/datasets.git 

->

```python
# a) How do flight prices vary with the days left until departure?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("flight_price.csv")

# Average price by days left
avg_price = df.groupby("days_left")["price"].mean()

# Plot
plt.figure(figsize=(8,5))
sns.lineplot(x=avg_price.index, y=avg_price.values, marker='o')
plt.title("Flight Price vs Days Left for Departure")
plt.xlabel("Days Left Before Departure")
plt.ylabel("Average Price (INR)")
plt.grid(True)
plt.show()
```

Explanation:

The graph shows that flight prices increase sharply as the departure date gets closer.

When more days are left, prices remain lower and stable, but in the last few days, they rise exponentially.

This means the best booking window is around 15–25 days before departure to get the lowest prices.

```python
# b) Compare prices across airlines for the same route (e.g., Delhi–Mumbai).
# Filter for Delhi to Mumbai route
route_df = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Average price by airline
avg_airline_price = route_df.groupby("airline")["price"].mean().sort_values()

# Plot
plt.figure(figsize=(9,5))
sns.barplot(x=avg_airline_price.values, y=avg_airline_price.index, palette="Blues_r")
plt.title("Average Flight Price by Airline (Delhi–Mumbai Route)")
plt.xlabel("Average Price (INR)")
plt.ylabel("Airline")
plt.show()
```

Explanation:


From the bar chart, we can see SpiceJet and AirAsia usually offer cheaper tickets, while Vistara and Air India are premium-priced airlines.


The difference in pricing is mainly due to service quality, timing flexibility, and in-flight facilities.


So, budget travelers can prefer AirAsia or SpiceJet, while comfort-focused travelers can choose Vistara or Air India.

Final Summary :

As the number of days left before departure decreases, flight prices go up very fast. This shows last-minute booking is costly.
Airlines like SpiceJet and AirAsia offer cheaper prices, while Air India and Vistara are premium airlines.
The ideal time to book is around 2–3 weeks before departure for best fares.

Q 10.  HR Analytics Dataset 
a). What factors most strongly correlate with employee attrition? Use visualizations to show key 
drivers (e.g., satisfaction, overtime, salary). 
b). Are employees with more projects more likely to leave?  
Dataset: hr_analytics

->

```python
# a) What factors affect employee attrition (leaving the company)?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("hr_analytics.csv")

# Check attrition by satisfaction level
plt.figure(figsize=(7,4))
sns.histplot(data=df, x="satisfaction_level", hue="left", kde=True, bins=20)
plt.title("Employee Attrition by Satisfaction Level")
plt.xlabel("Satisfaction Level")
plt.ylabel("Count")
plt.show()
```

Explanation:


The histogram shows that employees with low satisfaction levels are more likely to leave the company.

Higher satisfaction level → fewer employees leave.

So, job satisfaction is a major factor in attrition.

```python
# b) How does salary level influence employee retention?
# Count of employees left by salary level
salary_attrition = df.groupby("salary")["left"].mean().sort_values()

# Plot
plt.figure(figsize=(6,4))
sns.barplot(x=salary_attrition.index, y=salary_attrition.values, palette="coolwarm")
plt.title("Attrition Rate by Salary Level")
plt.xlabel("Salary Level")
plt.ylabel("Attrition Rate (Proportion)")
plt.show()
```

Explanation:


The chart clearly shows that employees with low salary have the highest attrition rate.

Those with medium and high salary are more stable and less likely to leave.

Therefore, salary and benefits are strong motivators for employee retention.

```python
# c) Department-wise attrition comparison
# Department-wise attrition
dept_attrition = df.groupby("sales")["left"].mean().sort_values()

# Plot
plt.figure(figsize=(8,4))
sns.barplot(x=dept_attrition.values, y=dept_attrition.index, palette="viridis")
plt.title("Attrition Rate by Department")
plt.xlabel("Attrition Rate")
plt.ylabel("Department")
plt.show()
```

Explanation:


Departments like sales and technical have slightly higher attrition.

Departments like HR and management show lower attrition.

This means that job pressure and workload differ across departments, affecting employee turnover.

Final Summary :

From the HR Analytics dataset, we can see that employee attrition mainly depends on satisfaction level, salary, and department.
Low satisfaction and low salary lead to high attrition.
Sales and technical teams show more attrition due to high workload.
To reduce attrition, companies should improve salary packages, work-life balance, and employee engagement.
